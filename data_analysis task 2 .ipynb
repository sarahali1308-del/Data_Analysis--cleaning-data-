```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('largest_companies.csv')
df.head()
```

```python
df.info()
```

```python
# 1. Remove the first row if it's just repeating the header
df = df.drop(0)

# 2. Convert Revenue and Profit to numbers (floats)
# We use errors='coerce' just in case there's a weird character
df['Revenue'] = pd.to_numeric(df['Revenue'], errors='coerce')
df['Profit'] = pd.to_numeric(df['Profit'], errors='coerce')

# 3. Check the types again
df.info()
```

```python
# Sort by revenue and take the top 10
top_10 = df.sort_values(by='Revenue', ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x='Revenue', y='Name', data=top_10, palette='viridis')

plt.title('Top 10 Global Companies by Revenue (in Billions USD)')
plt.xlabel('Revenue (USD Billions)')
plt.ylabel('Company Name')

plt.show()
```
